# **Definition (each attributes)**

* Impulsive (Real) is impulsiveness measured by BIS-11

```python
import pandas as pd 
import seaborn as sb
import matplotlib.pyplot as plt
import plotly.express as px 
import plotly.graph_objects as pg

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('Drug_Consumption.csv')
df
```

```python
#editing the data

df=df.rename(columns={'Amphet':'Amphetamine','Choc':'Chocolate','Caff':'Caffeine'})
df=df.rename(columns={"AScore":'Ascore'})
df[['Nscore','Escore','Oscore','Ascore','Cscore','Impulsive','SS']]=df[['Nscore','Escore','Oscore','Ascore','Cscore','Impulsive','SS']].astype(float)
```

# Question 1

Do education and drug's consumption correlate to each others?

```python
#list of education
Education=df['Education'].value_counts().index.tolist() 

#group education by CL6(use drugs in last days) and count CL6
new_df = df.copy()
new_df["CL6_count"] = (df.values == "CL6").sum(axis=1)


nnew_df = new_df[["CL6_count", "Education"]].groupby(["Education"]).sum()
Total= nnew_df.T.sum(axis=1)

fig=px.bar(nnew_df,y='CL6_count',title="Educations vs CL6").update_layout(xaxis_title="Education", yaxis_title="CL6")
fig.show()
```

ANS: People who still study in college or university, university degree, professional certificate participate in consuming drugs most respectively.

# Question 2
What is the most popular drugs (use in last day) among young adults (18-24 years old)?

```python
new_df=df[df['Age'] == '18-24'] 

#lists of drug
drugs_index=new_df.columns.tolist()[13:32] 
print(drugs_index)
#count the amount of consumption
drugs_f=new_df[drugs_index].apply(pd.value_counts)
drugs_f=drugs_f.T
drugs_f=drugs_f.fillna(0)

#sort by CL6(Used in last day)
print(drugs_f.sort_values(by=['CL6']))
```

ANS: Caffeine

# Question 3
What is the top 3 most popular drugs (used in last day) among all adults? Also make charts of it

```python
Age=df['Age'].value_counts().index.tolist() #lists of diffeerent age
a=[]

#make charts of different age VS CL6

for i in range(len(Age)):  
    new_df=df[df['Age'] == Age[i]]
    drugs_index=new_df.columns.tolist()[13:32]
    drugs_f=new_df[drugs_index].apply(pd.Series.value_counts)
    drugs_f=drugs_f.T
    drugs_f=drugs_f.fillna(0)
    drugs_f.sort_values(by=['CL6'])
    a.append(drugs_f['CL6'].tolist())

    fig=px.bar(drugs_f,y='CL6',title="Most popular drugs among adults' age between "+ Age[i]).update_layout(xaxis_title="Type of drugs", yaxis_title="CL6") 
    fig.show()

#make a chart of total age VS CL6

drugs_index=df.columns.tolist()[13:32]
all_drugs_f=df[drugs_index].apply(pd.Series.value_counts)
all_drugs_f=all_drugs_f.T
all_drugs_f=all_drugs_f.fillna(0)
fig=px.bar(all_drugs_f,y='CL6',title="Most popular drugs among all adults").update_layout(xaxis_title="Type of drugs", yaxis_title="CL6")
fig.show()

#make a chart of different age VSA CL6 in single chart

CL6=pd.DataFrame({'18-24':a[0],'25-34':a[1],'35-44':a[2],'45-54':a[3],'55-64':a[4],'65+':a[5],'total':all_drugs_f['CL6']}) 
fig=px.line(CL6,title="Most popular drugs among adults").update_layout(xaxis_title="Type of drugs", yaxis_title="CL6")
fig.show()

#make a table of CL6 of drugs VS different ages
print(CL6.sort_values(by=['total']))






```

ANS: Caffeine, Chocolate, Nicotine

# Question 4
Does personality changes when people gets older?

```python
Age=df['Age'].value_counts().index.tolist()

#USING plotly.graph_objects INSTEAD OF plotly.express (easier to make multiple lines in 1 chart)

#create score index
score_df=pd.DataFrame({'0': ['Nscore', 'Escore', 'Oscore', 'Ascore', 'Cscore','Impulsive','SS']})

#create data frame of mean of #score according to age.
for i in range (len(Age)) :
  score_df[i+1] =[df[df['Age']==Age[i]]['Nscore'].mean(),df[df['Age']==Age[i]]['Escore'].mean(),df[df['Age']==Age[i]]['Oscore'].mean(),
                  df[df['Age']==Age[i]]['Ascore'].mean(),df[df['Age']==Age[i]]['Cscore'].mean(),df[df['Age']==Age[i]]['Impulsive'].mean(),
                  df[df['Age']==Age[i]]['SS'].mean()]


#rename column
score_df=score_df.rename(columns={'0':'Score',1:'18-24',2:'25-34',3:'35-44',4:'45-54',5:'55-64',6:'65+'})


fig = pg.Figure()

fig.add_trace(pg.Scatterpolar(r=score_df['18-24'],theta=score_df['Score'],fill='toself',name='18-24'))
fig.add_trace(pg.Scatterpolar(r=score_df['25-34'],theta=score_df['Score'],fill='toself',name='25-34'))
fig.add_trace(pg.Scatterpolar(r=score_df['35-44'],theta=score_df['Score'],fill='toself',name='35-44'))
fig.add_trace(pg.Scatterpolar(r=score_df['45-54'],theta=score_df['Score'],fill='toself',name='45-54'))
fig.add_trace(pg.Scatterpolar(r=score_df['55-64'],theta=score_df['Score'],fill='toself',name='55-64'))
fig.add_trace(pg.Scatterpolar(r=score_df['65+'],theta=score_df['Score'],fill='toself',name='65+'))

fig.update_layout(width = 800,height = 800)
fig.show()
```

ANS: According to the chart above, yes.
